```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, PolynomialFeatures, LabelEncoder
```

* **Векторизация данных**

* **Выделение предсказываемого признака**

```python
x = data_set_work.drop(columns = 'popularity')
y = data_set_work['popularity']
features = x.columns.tolist()
comparison = pd.DataFrame()
comparison['metrics'] = ['mae','mse','rmse','r2', 'smape', 'rmsle']
```

*  **Функции расчета метрик**

```python
# Mean Absolute Error 
# показывает насколько в среднем прогнозы модели отклоняются от реальных значений по модулю
def _mae(y_true, y_pred):
    n = len(y_true)
    return 1 / n * np.sum(np.abs(y_true - y_pred))

# Mean Squared Error
# показывает насколько в среднем прогнозы модели отклоняются от реальных значений в квадрате
def _mse(y_true, y_pred):
    n = len(y_true)
    return 1 / n * np.sum((y_true - y_pred) ** 2)

# Коэффициент детерминации
# который показывает какая доля дисперсии целевых значений объясняется моделью
def _r2(y_true, y_pred):
    ssr = np.sum((y_true - y_pred) ** 2)
    sst = np.sum((y_true - y_true.mean()) ** 2)
    return 1 - ssr / sst
    
# Mean Absolute Percentage Error
# позволяет оценить в процентах насколько прогнозы модели отличаются относительно реальных значений
def _mape(y_true, y_pred):
    n = len(y_true)
    numerator = np.abs(y_true - y_pred)
    denominator = np.abs(y_true)
    return  1 / n * np.sum(numerator / denominator)
# Symmetric (симметричная) версия MAPE, подходит для для некоторых исключений 
def _smape(y_true, y_pred):
    n = len(y_true)
    numerator = 2 * np.abs(y_true - y_pred)
    denominator = (y_true + y_pred)
    return  1 / n * np.sum(numerator / denominator)

# Root Mean Squared Logarithmic Error
# позволяет оценить относительные ошибки в в логарифмическом масштабе
# Root (корень) помогает делать это относительно размерности исходных данных
def _rmsle(y_true, y_pred, c=1):
    n = len(y_true)
    return (1 / n * np.sum((np.log(y_true + c) - np.log(y_pred + c)) ** 2))**0.5
```

# Линейная регрессия

* **Нужно понять в каких диапазонах измеряются метрики и как на них влияют методы стандартизации.**

```python
from sklearn.linear_model import LinearRegression

# Методы стандартизации
scaler_std = StandardScaler()
x_std = scaler_std.fit_transform(x)
scaler_minmax =  MinMaxScaler()
x_minmax = scaler_minmax.fit_transform(x)
scaler_robust = RobustScaler()
x_robust = scaler_robust.fit_transform(x)
scalers = {'scaler_std': x_std ,'scaler_minmax': x_minmax ,'scaler_robust': x_robust }
# Таблица для результатов
results = pd.DataFrame()
results['metrics'] = ['mae','mse','rmse','r2','mape', 'smape', 'rmsle']

for scaler, x in scalers.items():
    # Разделение на обучающую и тестовую выборки
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    
    # Создание модели с заданными параметрами
    linear_reg = LinearRegression()
    
    # Обучение модели
    linear_reg.fit(X_train, y_train)
    
    # Предсказание на тестовой выборке
    y_pred = linear_reg.predict(X_test)

    # Оценки качества модели
    mae = _mae(y_test, y_pred)
    mse = _mse(y_test, y_pred)
    rmse = mse ** 0.5
    r2 = _r2(y_test, y_pred)
    mape = _mape(y_test, y_pred)
    smape = _smape(y_test, y_pred)
    rmsle =_rmsle(y_test, y_pred)
    results[scaler] = [mae, mse, rmse, r2, mape, smape , rmsle]

# Вывод оценок
results
```

***Значения метрик почти не меняются от выбора метода стандартизации (этого не видно из-за сокращения чисел), что довольно странно, но ошибки в коде у себя я не нашел***

* **Узнаем какие веса линейная регрессия дает различным признакам на обучающей и тестовой выборках c помощью Permutation Importance (метод перестановок), так как он универсален и мы сможем для следующих моделей использовать именно его, чтобы сравнивать.**

```python
from sklearn.inspection import permutation_importance
# Разделение на обучающую и тестовую выборки

X_train, X_test, y_train, y_test = train_test_split(x_std, y, test_size=0.2, random_state=42)

# Создание модели с заданными параметрами
linear_reg = LinearRegression()

# Обучение модели
linear_reg.fit(X_train, y_train)

#Расчет значимости признаков
weights = permutation_importance(linear_reg, x_std, y).importances_mean

sns.barplot(x=features,y= weights)
plt.title('Значимости различных признаков')
plt.xlabel('Признаки')
plt.xticks(rotation=90)
plt.ylabel('Значимость')
plt.show()
```

***Определяющими признаками в модели является сингловость,в меньшей степени инструментальность, кол-во жанров и кол-во речи.***

* **Унаем как меняются оценки метрик при крос-валидации на каждом из фолдов.**
